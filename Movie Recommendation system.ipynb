```python
#it is in format of list of dictionaries
#so we need to convert into:-
#['Action', 'Adventure', 'Fantasy', 'Science Fiction']

#but firstly we need to convert string of list to list
#to convert this, python has a module called ast
#ast module mai ek function hota hai - literal_eval
```

```python
import ast
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'].apply(convert)       #is used in Pandas to apply a function to every value of the genres column.
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
#create a stemming tool in Python using NLTK.
#This line imports the Porter Stemmer algorithm from the NLTK library.
#Porter Stemmer is used to reduce words to their root form(stem) in text processing.
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split(): 
        y.append(ps.stem(i))

    return " ".join(y)          #convert string into list 
```

```python
new_df['tags'] = new_df['tags'].apply(stem)       #is used to apply the stemming function to every row in the tags column.
```

```python
#we calculate distance from one movie to another movie

#Euclidean distance is the straight-line distance between two points.
#Cosine similarity measures the angle between two vectors, not the distance.

#Jitna high dimensional data(5000) mai kaam karoge utna hi euclidean distance utna fail karta hai
#so we use cosine distance 

#similarity is like inverse of distance, always in b/w 0 and 1
#1 - high similarity 
#0 - low similarity 
```

```python
#imports the cosine similarity function from the scikit-learn library.
#Cosine similarity checks the angle between two vectors.
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]

    for i in movies_list: 
        print(i[0])
```
